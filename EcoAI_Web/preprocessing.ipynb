```python
def replacing_shortcut(columnname, dataset):
    x = dataset[columnname].value_counts()
    fields = x.keys()
    replacers = [i for i in range(len(fields))]

    dic = dict()
    for i in range(len(fields)):
        dic[str(fields[i])] = replacers[i]
    dataset = dataset.replace({columnname:dic})
    print(x)
    print(fields)
    return dataset
```

```python
dataset = replacing_shortcut("City", dataset)
```
